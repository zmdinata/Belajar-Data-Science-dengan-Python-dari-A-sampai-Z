```python
import pandas as pd
```

```python
iris = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
```

```python
iris = iris.rename(columns = {"variety": "species"}) #kalo mau commit harus ada iris=
```

```python
iris.head(10)
```

```python
iris.info()
```

```python
#Value Count
iris["species"].value_counts()
```

```python
iris["petal.length"].value_counts() #ga cocok kalo itung yg numerik, cocoknya object saja
```

```python
#Group By, 1 aggregation

iris.head()
```

```python
iris.groupby(["species"]).count()
```

```python
iris.groupby(["species"]).mean()
```

```python
iris.groupby(["species"]).median()
```

```python
iris.groupby(["species"]).min()
```

```python
iris.groupby(["species"]).max()
```

```python
iris.groupby(["species"]).sum()
```

```python
iris.groupby(["species"]).describe().T
```

```python
df = iris.groupby(["species"]).mean() #species bukan kolom disini tapi index
df
```

```python
df = iris.groupby(["species"]).mean().reset_index() #jadinya kita reset index dan species jadi kolom
df
```

```python
#Group By, 2 or more aggregation
iris.head()
```

```python
iris.groupby(["species"]).agg(["min", "max"]) #jadi ada 2 agg
```

```python
iris.groupby(["species"]).agg({"sepal.length": "min", "sepal.width": "max"}) #kalo beda beda aggresinya
```

```python
iris.groupby(["species"]).agg({"sepal.length": ["min", "sum"], "sepal.width": ["max", "mean"]})
```
