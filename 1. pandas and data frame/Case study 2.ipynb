```python
import pandas as pd
import numpy as np
```

```python
order_history = pd.DataFrame({
    'order_id': [1,2,3,4,5,6,7,8],
    'cust_id': ['c1','c2','c3','c5','c6','c7','c8','c10'],
    'order_amount': [100,200,250,150,300,350,100,150]
})

order_history
```

```python
customer_base = pd.DataFrame({
    'cust_id': ['c1','c2','c3','c4','c5','c6','c7','c8','c9','c10'],
    'city': ["Jakarta", "Tangerang", "Jakarta", "Jakarta", "Bandung", "Tangerang", "Jakarta","Bandung","Tangerang","Tangerang"]
})

customer_base
```

```python
df_joinned = order_history.merge(customer_base, how = "inner", left_on = "cust_id", right_on = "cust_id")
df_joinned 
```

```python
df_grouped = df_joinned.groupby("city").agg({"order_amount": "sum"}).reset_index()
```

```python
# Sekarang kode ini akan berjalan karena df_grouped adalah DataFrame
df_grouped.columns = ["city", "total_order_amount"]
```

```python
df_grouped
```

```python
df_grouped.sort_values(by = "total_order_amount", ascending = False).reset_index(drop = True)
```
