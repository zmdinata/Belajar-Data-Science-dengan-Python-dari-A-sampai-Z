```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'cust_id': ['c1','c2','c3','c4','c5'],
    'item_bought': [10,15,20,35,4],
    'city': ["Bandung", "Jakarta", "Bogor", "Tangerang", "Bali"]
})

df1
```

```python
df3 = pd.DataFrame({
    'cust_no': ['c1','c2','c3', 'c5','c6','c7','c8'],
    'age': [24,31,23,40,35,28,34]
})

df3
```

```python
#merge
df1.merge(df3, how = "left", left_on = "cust_id", right_on = "cust_no") #left join
```

```python
df1.merge(df3, how = "right", left_on = "cust_id", right_on = "cust_no")  #right join
```

```python
df1.merge(df3, how = "inner", left_on = "cust_id", right_on = "cust_no") #inner join
```

```python
df1.merge(df3, how = "outer", left_on = "cust_id", right_on = "cust_no") #full join (nama parameternya how = "outer"
```

```python
df1.merge(df3, how = "inner", left_on = "cust_id", right_on = "cust_no").drop(columns = ["cust_no"]) #menghapus column
```
