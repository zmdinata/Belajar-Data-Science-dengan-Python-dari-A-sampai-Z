```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({ 'A' : [1,2,3,4],
                  'B' : [10,20,30,40],
                  'C' : [20,40,60,80]})
```

```python
df
```

```python
#APPLY

def sum(x):
    result = x.sum()
    return result
df['d'] = df.apply(sum, axis = 1) #columns apply (A,B,C)
```

```python
df
```

```python
df.loc[4,:] = df.apply(sum, axis = 0) #Row apply
```

```python
df
```

```python
def multiply_by_2(x):
    return x*2
df["E"] = df["B"].apply(multiply_by_2)
```

```python
df
```
