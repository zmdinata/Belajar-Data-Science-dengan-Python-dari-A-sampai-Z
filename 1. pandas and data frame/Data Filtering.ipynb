```python
#Negative of a criteria
df
```

```python
df[(df["item_bought"] > 10)]
```

```python
df[~(df["item_bought"] > 10)] #kebalikan(tidakdi atas 10)/ logika negasi
```

```python
#Filter df w/ several criterias (AND)

df
```

```python
df[(df["item_bought"] > 10) & (df["city"] != "Bogor")]
```

```python
df[~((df["item_bought"] > 10) & (df["city"] != "Bogor"))]
```

```python
#Filter df w/ several criterias (OR)

df[(df["item_bought"] > 10) | (df["city"] != "Bogor")]
```

```python
df[~((df["item_bought"] > 10) | (df["city"] != "Bogor"))]
```

```python
#Filter w/ isin operation

list_of_cities = ["Jakarta", "Tangerang", "Depok"]
```

```python
df[(df["city"].isin(list_of_cities))]
```
