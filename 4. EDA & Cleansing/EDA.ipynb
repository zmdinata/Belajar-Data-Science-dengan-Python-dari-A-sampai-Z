```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Univariate EDA

```python
df = pd.read_csv('train.csv')
df
```

```python
df.info()
```

```python
## Descriptive Statistics

df.describe()
```

```python
##BoxPlot
```

```python
num_dtypes = ["int32", "int64","float32","float64"]

df_num = df.select_dtypes(num_dtypes) #ngambil semua kolom yang isinya numerik
num_col = df_num.columns
num_col
```

```python
plt.figure(figsize = (12,4))
for i in range(len(num_col)):
    plt.subplot(1, len(num_col), i+1)
    sns.boxplot(y = num_col[i], data = df, orient = "v")
    plt.tight_layout()
```

```python
##Histogram
```

```python
plt.figure(figsize = (15,5))
for i in range(len(num_col)):
    plt.subplot(1, len(num_col), i+1)
    sns.histplot(x = num_col[i], data = df)
    plt.tight_layout()
```
