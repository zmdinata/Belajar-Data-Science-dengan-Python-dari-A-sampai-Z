### Additional Data Preparation Before Modeling

```python
import pandas as pd
```

```python
df = pd.read_csv("Clean_Dataset.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df_copy = df.copy()
df_copy
```

```python
print(df_copy.columns)
```

```python
df_copy.columns = df_copy.columns.str.strip()
```

```python
df_copy_drop = df_copy.drop(["Unnamed: 0", "flight"], axis=1)
df_copy_drop
```

```python
## One Hot Encoding
```

```python
df_obj = df_copy_drop.select_dtypes(["object"])
col_obj = df_obj.columns
```

```python
col_obj
```

```python
df_clean = pd.get_dummies(df_copy_drop, columns = col_obj)
df_clean
```

```python
##Train Test Split
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_clean["price"]
x = df_clean.drop(columns = ["price"])
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Regression

```python
from sklearn import linear_model
```

```python
linear_reg = linear_model.LinearRegression()

linear_reg.fit(x_train, y_train)
```

```python
print(linear_reg.coef_)
```

```python
print(linear_reg.intercept_)
```

```python
## Evaluate
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score
```

```python
y_pred = linear_reg.predict(x_test)
y_pred
```

```python
mean_absolute_error(y_test, y_pred) #kalo dibawah setengah Standar deviasi berarti model kita cukup akurat
```

```python
root_mean_squared_error(y_test, y_pred) #kalo dibawah setengah Standar deviasi berarti model kita cukup akurat
```

```python
 import numpy as np
```

```python
np.mean(y_test)
```

```python
r2_score(y_test, y_pred) #bahwa error kita kecil karena akurasi dari dataset ini 91%
```
