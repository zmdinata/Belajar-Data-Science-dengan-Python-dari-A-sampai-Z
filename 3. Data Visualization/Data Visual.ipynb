```python
# Menghapus spasi di awal dan akhir dari semua nama kolom
df.columns = df.columns.str.strip()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
dfg = df.groupby("Sales Channel").agg({"Total Profit": "sum"}).reset_index()
dfg
```

```python
# Group by 'Order Date' lalu jumlahkan 'Total Profit' dan 'Total Revenue'
df_harian = df.groupby('Order Date').agg({
    'Total Profit': 'sum',
    'Total Revenue': 'sum'
}).reset_index()

# Urutkan berdasarkan tanggal untuk memastikan garis plot tidak kacau
df_harian = df_harian.sort_values(by='Order Date')
```

COMPARISON

```python
plt.bar(x = "Sales Channel", height = "Total Profit", data = dfg)
```

```python
sns.barplot(x = "Sales Channel", y = "Total Profit", data = dfg)
```

```python
 ## Comparison _ Multiline chart
```

```python
plt.figure(figsize = (12,6))
plt.xticks(rotation = 90)

sns.lineplot(x="Order Date", y="Total Profit", data=df_harian, label="Total Profit")
sns.lineplot(x="Order Date", y="Total Revenue", data=df_harian, label="Total Revenue")

plt.legend()
```

```python
# Ganti 'figure.size' menjadi 'figure.figsize'
sns.set_theme(rc={'figure.figsize': (10, 5)})
sns.set(font_scale=0.5)

sns.lineplot(x="Order Date", y="Total Profit", data=df_harian, label="Total Profit")
sns.lineplot(x="Order Date", y="Total Revenue", data=df_harian, label="Total Revenue")

plt.legend()
plt.show() # Jangan lupa plt.show() untuk menampilkan plot
```
